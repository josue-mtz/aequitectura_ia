```python
def relu(x):
    """
    Implementación manual de la función ReLU (Rectified Linear Unit).
    
    ReLU es una función de activación muy común en redes neuronales profundas.
    Funciona así:
    - Si x > 0: devuelve x (mantiene el valor original)
    - Si x <= 0: devuelve 0 (anula valores negativos)
    
    En fórmula matemática: ReLU(x) = max(0, x)
    
    Args:
        x (int o float): Valor numérico a procesar
    
    Returns:
        int o float: El valor máximo entre 0 y x
    
    Raises:
        TypeError: Si x no es un número (int o float)
    """
    if not isinstance(x, (int, float)):
        raise TypeError("El valor debe ser numérico (int o float).")
    return max(0, x)

# EJEMPLO DE USO
# Creamos una lista con diferentes valores (positivos, negativos y cero)
valores = [-5, -0.5, 0, 2.3, 10]

# Aplicamos la función ReLU a cada valor usando list comprehension
resultado = [relu(v) for v in valores]

# Mostramos los resultados
print("Entrada: ", valores)
print("Salida ReLU: ", resultado)
```

Interpretación:
- Los valores negativos (-5, -0.5) se convierten a 0
- El cero se mantiene igual (0)
- Los valores positivos (2.3, 10) se mantienen igual

```python
# VISUALIZACIÓN DE LA FUNCIÓN RELU
# Este código crea una gráfica para visualizar cómo funciona la función ReLU

import matplotlib.pyplot as plt

# Crear figura con tamaño específico (ancho=10, alto=6 pulgadas)
plt.figure(figsize=(10, 6))

""" Graficar los puntos:
 - valores: eje X (valores de entrada)
 - resultado: eje Y (valores de salida de ReLU)
 - marker='o': muestra puntos circulares en cada dato
 - linewidth=2: grosor de la línea que conecta los puntos
 - markersize=8: tamaño de los círculos
"""
plt.plot(valores, resultado, marker='o', linewidth=2, markersize=8, label='ReLU(x)')

# Añadir línea horizontal en y=0 (línea roja punteada)
# Esta línea ayuda a visualizar donde ocurre el "corte" de ReLU
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5, label='Umbral (y=0)')

# Etiquetas y título
plt.xlabel('Valores de Entrada (x)', fontsize=12)
plt.ylabel('Salida ReLU(x)', fontsize=12)
plt.title('Función ReLU - Comportamiento Visual', fontsize=14, fontweight='bold')

# Mostrar cuadrícula para facilitar lectura
plt.grid(True, alpha=0.3)

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()

print("✓ Gráfica generada exitosamente")
```

Observaciones:
- La gráfica muestra cómo ReLU 'corta' todos los valores negativos a 0
- Todos los valores negativos caen en la línea roja horizontal (y=0)
- Los valores positivos se mantienen sin cambios
